```python
uniq_ops = {}
for col_name in list(aDF.columns):
    if col_name in ["Time of Day","Process Name","PID"]:  #skip these
        continue
    _tmp = aDF.loc[:,col_name].unique()
    print(f"{col_name}: {_tmp}")
    uniq_ops[col_name] = _tmp
print(uniq_ops.keys())
```
