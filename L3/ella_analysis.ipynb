```python
import numpy as np
import pandas as pd
```

```python
ellaDF=pd.read_csv("ella_procmon_log.csv")
```

```python
ellaDF.head(5)
```

```python
col_names= list(ellaDF.columns)
col_names
```

```python
uniq_ops = {}
for col_name in col_names:
    if col_name in ["Time of Day","Process Name","PID"]:  #skip these
        continue
    _tmp = ellaDF.loc[:,col_name].unique()
    print(f"{col_name}: {_tmp}")
    uniq_ops[col_name] = _tmp
```

```python
uniq_ops["Operation"]
```

```python
RegEvents = ellaDF[(ellaDF["Operation"] == 'RegCreateKey') | (ellaDF["Operation"] ==  'RegSetValue')]
for i in range(len(RegEvents)):
    print(f"{RegEvents.iloc[i,3]} -- {RegEvents.iloc[i,4]} -- {RegEvents.iloc[i,6]}")
```
